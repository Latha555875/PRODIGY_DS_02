```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x="/home/pulicherla/Documents/test.xlsx"
df=pd.read_excel(x)
print(df.head())


# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# viualise the "age" data in bargraphs

age_counts=df["Age"].value_counts().sort_index()
axes[0].bar(age_counts.index, age_counts.values,color='brown')
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Visualise the age data using bargraphs")

# Drop missing values in 'Age' column for clean visualization
age_data = df['Age'].dropna()

# Plotting the histogram
axes[1].hist(age_data, bins=20, color='skyblue')
axes[1].set_title('Visualise the age data using histograms')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
file = pd.read_excel('/home/pulicherla/Documents/test.xlsx')
print(file.head())


# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns


# Scatter Plot: Age vs. Fare

sns.scatterplot(x='Age', y='Fare', hue='Survived', data=file, palette='dark',ax=axes[0])
axes[0].set_title('Scatter Plot of Age vs Fare')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Fare')

# Box plot for Pclass and survived

sns.barplot(x='Age', y='Fare', hue='Survived', data=file,ax=axes[1])
axes[1].set_title('Bar plot of Age vs Fare')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Fare')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
file = pd.read_excel('/home/pulicherla/Documents/test.xlsx')
print(file.head())

# Pairplot
sns.pairplot(file[['Age', 'Fare', 'Pclass','Survived']], hue='Survived' ,palette='dark')
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()

fig, axes = plt.subplots(1, 1, figsize=(10, 6))  # 1 row, 1 column

# Violin Plot: Fare by Pclass

sns.violinplot(x='Pclass', y='Fare',hue='Survived' , data=file, palette='dark', split=True, ax=axes)
axes.set_title('Violin Plot of Fare by Pclass')
axes.set_xlabel('Pclass')
axes.set_ylabel('Fare')
plt.tight_layout()
plt.show()
```
